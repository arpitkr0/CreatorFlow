```python
import pandas as pd
import glob
import os

csv_files = glob.glob('*.csv')

all_data = []


for file in csv_files:
    try:
        df = pd.read_csv(file)

        all_data.append(df)
        print(f"Successfully read: {file}")
    except Exception as e:
        print(f"Error reading {file}: {e}")

if all_data:
    concatenated_df = pd.concat(all_data, ignore_index=True)
    concatenated_df.to_csv('concatenated_dataset.csv', index=False)
    print("Successfully concatenated all CSV files into 'concatenated_dataset.csv'")
else:
    print("No CSV files found in the current directory.")
```

```python
import pandas as pd

try:
    concatenated_df = pd.read_csv('concatenated_dataset.csv')
except FileNotFoundError:
    print("Error: 'concatenated_dataset.csv' not found. Please run the concatenation script first.")
    exit()

shuffled_df = concatenated_df.sample(frac=1, random_state=42).reset_index(drop=True)

shuffled_df.to_csv('shuffled_dataset.csv', index=False)

print("Successfully shuffled the rows and saved to 'shuffled_dataset.csv'")
```
